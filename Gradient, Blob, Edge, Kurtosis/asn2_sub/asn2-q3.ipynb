```python
# boilerplate imports

import matplotlib.pyplot as plt
import skimage.data as skdata
import skimage.feature as skfeat
import skimage.color as skcolor
import skimage.io as skio
```

# Step 1: Perform Canny edge detection on the `skdata.astronaut()` image.

- call skimage.astronaut() to obtain the image
- convert it to grayscale
- perform Canny edge detection (`skfeat.canny()` function) on the grayscale image -- select the parameters to optimize the edge detection to find only the edges of major salient objects in the scene such as the person, helmet, flag, and space shuttle model (you won't be able to do this perfectly, but do the best you can).  
- display the result of the Canny edge detection as a binary image.

```python
# Write your code here.
import skimage.util as util
I = skcolor.rgb2gray(util.img_as_float(skdata.astronaut()))


plt.subplots(1,2,figsize=(25,25))
plt.subplot(1,2,1)
plt.imshow(I, cmap='gray', vmin=0, vmax=1)

plt.subplot(1,2,2)

edges = skfeat.canny(I, low_threshold=.4, high_threshold=.5, sigma=1)
plt.imshow(edges, cmap='gray')
```

# Step 2: Justify your parameter selection

Explain why you selected the parameters that you did and justify why they are appropriate for the given edge detection task and context.

*Write your answer here:*
I set the  low threshold to 0.4 to allow the algorithm to pick up finer details of the edges without capturing excessive noise. A lower threshold might result in too many non-important edges, while a higher threshold might miss important details.

The high threshold is set higher 0.5 to filter out weak gradients that are likely noise. This ensures that only the stronger, more significant edges of major objects like the astronaut’s helmet and flag are detected.

I chose 1 to slightly smooth the image, reducing noise without overly blurring the important edges.
